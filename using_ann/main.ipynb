```python
import import_ipynb
from utils.dataset import jobDataset
from model.model import jobModel
import pandas as pd
from torch.utils.data import DataLoader
import torch.nn.functional as F
import torch

BATCH = 50
EPOCHS = 200

def train(mode, model):
    train_data=jobDataset(mode)
    train_loader=DataLoader(train_data, batch_size=BATCH, shuffle=True)
    print("Train Data Loading Success! Size is {}.".format(len(train_loader.dataset)))
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-5)
    model.train()

    for epoch in range(EPOCHS):
        for batch_idx, samples in enumerate(train_loader):
            x_train, y_train=samples  ## x_train size : [bs, 4]
            print(x_train.shape)
            logits = model(x_train)
            #print(logits)
            loss=F.binary_cross_entropy(logits.squeeze(), y_train)
            loss.retain_grad()
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            print('Epoch {:4d}/{} Batch {}/{} Loss: {:.6f}'.format(epoch+1, EPOCHS, batch_idx+1,len(train_loader), loss.item()))
    print("Train Finished")
  
def test(mode, model):
    test_data=jobDataset(mode)
    test_loader=DataLoader(test_data, batch_size=BATCH, shuffle=True)
    print("Test Data Loading Success! Size is {}.".format(len(test_loader.dataset)))
    model.eval()
    predict=[]

    for batch_idx, samples in enumerate(test_loader):
        x_test=samples  ## x_train size : [bs, 21]
        logits = model(x_test)
        logits=logits.view(-1)
        #import pdb;pdb.set_trace()
        for t in logits:
            if t>=0.5: predict.append(1)
            else: predict.append(0)
        
    submission = pd.DataFrame(predict, columns=['applied'])
    submission.to_csv('./ann_submission.csv', index=False)
    print("Test Finished")


model = jobModel(711,1)  ### 4 
train("Train",model)
test("Test", model)
torch.save(model.state_dict(),"./weight/train_model.dat")


```
